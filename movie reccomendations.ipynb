```python
movies=movies[['id', 'title', 'genre', 'overview']] 
```

```python
movies
```

```python
movies['tags']= movies['overview']+movies['genre']
```

```python
movies
```

```python
new_data =movies.drop(columns=['overview','genre'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="Schindler's List"].index[0]
```

```python
new_data.info()
```

```python
distance=sorted(list(enumerate(similarity[3])),reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(i)
```

```python
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def reccomendation(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance=sorted(list(enumerate(similarity[index])),reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
reccomendation('Iron Man')
```
